```python
'''
python常见的字符串操作
'''

# 创建字符串,可以使用单引号、双引号、三引号
s1 = 'Hello'
s2 = "World"
s3 = '''This is a
multi-line string'''

# 访问字符串
print(s1)
print(s2[-1])
print(s3[5:10])

# 字符串连接，可以使用加号拼接、使用乘号重复
s4 = s1 + ' ' + s2
s5 = s1 * 3
print(s4)
print(s5)

# 两种常见的格式化字符串方法
name = 'Alice'
age = 18
# 使用format方法
s6 = 'My name is {}, I am {} years old.'.format(name, age)
# 使用f-string
s7 = f'My name is {name}, I am {age} years old.'
print(s6)
print(s7)

# 转换字符串
n = 3.14
s8 = str(n)
# repr方法，返回对象的string格式
s9 = repr(n)
'''
repr()的目的是让解释器或开发者能够准确地知道对象的类型和内容，它的返回结果通常包含类名和属性信息，而且可以用eval()函数还原对象。
str()的目的是让用户能够方便地阅读对象的信息，它的返回结果通常更加简洁和可读。
当使用print()函数或者格式化字符串时，会优先调用str()方法；当直接在交互式窗口输出对象时，会调用repr()方法。
如果一个类没有定义str()方法，那么会默认使用repr()方法来转换对象为字符串。
'''
print(s8) # 输出3.14
print(s9) # 输出3.14

# 比较字符串
s10 = 'apple'
s11 = 'banana'
print(s10 == s11)
print(s10 != s11)
'''
Python中，大于号（>）和小于号（<）用于两个字符串之间的含义是比较它们的字典序。
也就是说，先比较两个字符串的第一个字符的Unicode编码，如果相等则继续比较后续字符，直到找到不相等的字符或者其中一个字符串结束。
如果不相等的字符的Unicode编码较小，则该字符串小于另一个字符串；
如果其中一个字符串结束，则该字符串小于另一个字符串。
可以使用Python内置的ord()函数来查看每个字符的Unicode编码。
'''
print(f'"abc" < "abcd": {"abc" < "abcd"}')
print(f'"abce" > "abcdf": {"abc" < "abcd"}, 因为ord("e") > ord("d")=>{ord("e")} > {ord("d")}')

# 查找和替换
s12 = 'Hello World'
# 输出4
print(s12.find('o'))
# 输出4
print(s12.index('o'))
# 输出-1
print(s12.find('x'))
# 抛出异常
# print(s12.index('x'))
# 输出Hello Python
print(s12.replace('World', 'Python'))

# 分割和合并
s13 = 'red,green,blue'
s14 = 'one,two,three,four'
# 输出['red', 'green', 'blue']
print(s13.split(','))
'''
partition()方法是用来根据指定的分隔符将字符串进行分割的。
如果字符串包含指定的分隔符，则返回一个3元的元组：
第一个为分隔符前面的子字符串，
第二个为分隔符本身，
第三个为分隔符后面的子字符串。
如果字符串不包含指定的分隔符，则返回一个元组，第一个为整个字符串，第二个和第三个为空字符串。
'''
# 输出('one', ',', 'two,three,four')
print(s14.partition(','))
print(s14.partition(';'))
s15 = ['a', 'b', 'c']
# 使用-连接
s16 = '-'.join(s15)
# 输出a-b-c
print(s16)

# 大小写转换
s17 = 'hello iPhone18maxPro'
s18 = 'HELLO WORLD'
# 输出HELLO IPHONE18MAXPRO
print(s17.upper())
# 输出hello world
print(s18.lower())
'''
capitalize()方法只将字符串的第一个字母变成大写，其他字母变小写。
如果字符串的第一个字符不是字母，那么它不会改变。
'''
# 输出Hello world
print(s18.capitalize())
'''
title()方法将字符串内的所有单词的首字母变成大写，其他字母变小写。
如果字符串中有非字母的字符，那么它会将其后的第一个字母视为单词的开头。
'''
# 输出Hello Iphone18Maxpro
print(s17.title())

# 去除空白
s19 = '  hello  '
s20 = '\nworld\n'
# 输出hello
print(s19.strip())
# 输出world\n
print(s20.lstrip())
# 输出\nworld
print(s20.rstrip())

# 判断属性
s21 = 'abc123'
s22 = 'abc'
s23 = '123'
'''
isalnum()方法是用来检测字符串是否由字母和数字组成的。
如果字符串至少有一个字符并且所有字符都是字母或数字则返回True，否则返回False。
'''
# 输出True
print(s21.isalnum())
# 判断是否为字母字符串，输出True
print(s22.isalpha())
# 判断是否为数字字符串，输出True
print(s23.isdigit())

# 检查子串
s24 = 'Hello World'
# 判断字符串包含，输出True
print('Hello' in s24)
# 判断字符串前缀，输出True
print(s24.startswith('He'))
# 判断字符串后缀，输出True
print(s24.endswith('ld'))

# 填充字符串
s25 = 'abc'
'''
ljust()方法是用来将字符串左对齐，并使用指定的字符（默认为空格）填充至指定的长度的。
如果指定的长度小于字符串的长度，则返回原字符串。
rjust()方法是右对齐，center()方法是中间对齐，用法和ljust()一致。
'''
print(s25.ljust(10, '*')) # 输出abc*******
print(s25.rjust(10, '*')) # 输出*******abc
print(s25.center(10, '*')) # 输出***abc****

# 编码和解码
s26 = '你好'
# 使用utf-8编码，编码后的字符串是bytes类型的，可以用于网络传输或存储。
b = s26.encode('utf-8')
# 输出b'\xe4\xbd\xa0\xe5\xa5\xbd'
print(b)
# 使用utf-8解码
s27 = b.decode('utf-8')
# 输出 你好
print(s27)

# 转义和反转义
from html import escape, unescape # 导入html模块中的escape和unescape函数
s28 = '<p>Hello World</p>'
# 转义特殊字符
s29 = escape(s28)
# 输出&lt;p&gt;Hello World&lt;/p&gt;
print(s29)
# 反转义特殊字符
s30 = unescape(s29)
# 输出<p>Hello World</p>
print(s30)
```
