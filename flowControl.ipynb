Python中流程控制的实现方式主要有三种，分别是选择结构、循环结构和跳转结构。
* 选择结构是根据条件的真假来执行不同的代码，常用的语句有if、elif和else。
* 循环结构是根据条件或者可迭代对象来重复执行一段代码，常用的语句有while和for。
* 跳转结构是用来改变程序的执行流程，常用的语句有break、continue和pass。

```python
# 条件语句根据一个或多个条件的真假来执行不同的代码块。
'''
Python中有四种条件语句：
if语句；
if-else语句；
if-elif-else语句；
嵌套if-else语句。
'''
print('-' * 18, "条件语句", '-' * 18)
# if语句
x = 1
if x > 0:
  print(f"{x}是正数")

# if-else语句
y = -1
if y > 0:
  print(f"{y}是正数")
else:
  print(f"{y}是负数")

# if-elif-else语句
z = 0
if z > 0:
  print(f"{z}是正数")
elif z < 0:
  print(f"{z}是负数")
else:
  print(f"{z}是零")

# 嵌套if-else语句
a = 1
b = 2
if (a > b):
  print(f"{a}大于{b}")
else:
  if a < b:
    print(f"{a}小于{b}")
  else:
    print(f"{a}等于{b}")


# 循环语句用于重复执行一个或多个代码块，直到满足某个条件。
'''
Python中有两种循环语句：
for循环；
while循环。
'''
print('-' * 18, "循环语句", '-' * 18)
# for循环
words = ["Hello", "World", "Python"]
for word in words:
  print(word)

# while循环
n = 1
while n < 3:
  print(n)
  n = n + 1


# 转移语句用于改变程序的执行流程，跳出或继续某个循环。
'''
Python中有三种转移语句：
break语句；
continue语句；
pass语句。
'''
print('-' * 18, "转移语句", '-' * 18)
# break语句
for i in range(3):
    if i == 5:
        break # 跳出循环
    print(i)

# continue语句
for j in range(3):
    if j % 2 == 0:
        continue # 跳过本次循环
    print(j)

# pass语句
def func():
    pass # 占位符，不做任何操作
```
