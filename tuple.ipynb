元组支持的常见操作
* 创建：可以使用圆括号（）来创建一个元组，也可以使用tuple()构造函数来创建一个元组，或者直接将一系列值用逗号分隔，不用括号。
* 访问：可以使用索引（[]）或切片（[:]）来访问元组中的元素，索引从0开始，也可以是负数，表示从后往前数。
* 修改：元组是不可变的，意味着不能修改、添加或删除元组中的元素，但是可以通过连接（+）或重复（*）操作来创建一个新的元组。
* 删除：不能删除元组中的元素，但是可以使用del语句来删除整个元组。
* 连接：可以使用+运算符来连接两个或多个元组，得到一个新的元组。
* 重复：可以使用*运算符来重复一个元组若干次，得到一个新的元组。
* 成员检查：可以使用in或not in运算符来检查一个值是否在一个元组中，返回True或False。
* 迭代：可以使用for循环来遍历一个元组中的所有元素。


Python为元组提供了两个内置方法，count()和index()。
* count(x)用于返回x在元组中出现的次数；
* index(x)用于返回x在元组中第一次出现的索引位置。

```python
print('-' * 18, "创建元组", '-' * 18)
# 创建一个元组
colors = ("red", "green", "blue")
print(colors)

# 使用tuple()构造函数创建一个元组
numbers = tuple((1, 2, 3))
print(numbers)

# 直接用逗号分隔值创建一个单个值的元组
# 注意要加逗号，否则会被当成普通的括号
single = ("hello",)
print(single)

print('-' * 18, "访问元组", '-' * 18)
# 访问元组中的第二个元素
print(colors[1])
# 访问元组中的最后一个元素
print(colors[-1])
# 访问元组中的一部分元素
print(colors[0:2])

print('-' * 18, "修改元组", '-' * 18)
# 报错 TypeError: 'tuple' object does not support item assignment
# colors[0] = "pink"
# 创建一个新的包含"pink"的元组
new_colors = colors + ("pink",)
print(new_colors)
# 创建一个新的重复两次的元组
double_colors = colors * 2
print(double_colors)

print('-' * 18, "删除元组", '-' * 18)
del colors
# 报错 NameError: name 'colors' is not defined
# print(colors) 

print('-' * 18, "连接元组", '-' * 18)
fruits = ("apple", "banana", "cherry")
animals = ("cat", "dog", "bird")
mix = fruits + animals
print(mix)

print('-' * 18, "重复元组", '-' * 18)
repeat = fruits * 3
print(repeat)

print('-' * 18, "成员检查", '-' * 18)
print("apple" in fruits)
print("orange" not in fruits)

print('-' * 18, "遍历元组", '-' * 18)
for x in animals:
  print(x)
print('-' * 18, "内置方法", '-' * 18)
print(f'animals.count("cat"): {animals.count("cat")}')
print(f'animals.index("cat"): {animals.index("cat")}')
```
