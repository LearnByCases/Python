```python
'''
Python中，{}、[]、()的用法
'''

# Python的风格指南PEP 8建议避免在if、for、while等语句中使用不必要的小括号。
# 实测在for循环后面加小括号，在jupyter中会报错。

'''
大括号{}用于创建字典或集合。
'''
print('-' * 18, "大括号{}用于创建字典或集合", '-' * 18)
# 字典是一种映射类型，用于存储键值对，键必须是唯一的和可哈希的。
# 创建一个字典
d = {"name": "Alice", "age": 20, "gender": "female"}
# 访问字典中的值
print(d["name"])
# 修改字典中的值
d["age"] = 21
# 添加新的键值对
d["city"] = "Beijing"
# 删除键值对
del d["gender"]

# 集合是一种无序的，不重复的元素集合，支持集合运算。
# 创建一个集合
s = {1, 2, 3, 4, 5}
# 访问集合中的元素
for x in s:
    print(x)
# 添加元素到集合
s.add(6)
# 删除元素从集合
s.remove(1)
# 求两个集合的交集
t = {3, 4, 5, 6, 7}
print(s & t)


'''
中括号[]用于创建列表或索引。
'''
print('-' * 18, "中括号[]用于创建列表或索引", '-' * 18)
# 列表是一种有序的，可变的序列类型，用于存储任意类型的元素。
# 创建一个列表
l = [1, 2, 3, "a", "b", "c"]
# 访问列表中的元素
print(l[0])
print(l[-1])
# 修改列表中的元素
l[0] = 10
# 添加元素到列表末尾
l.append(4)
# 插入元素到列表指定位置
l.insert(3, "d")
# 删除列表中的元素
l.pop() # 返回并删除最后一个元素
l.remove("a") # 删除第一个出现的"a"

# 索引是一种访问序列类型（如字符串，列表，元组）中某个位置或范围的元素的方法。
# 创建一个字符串
s = "Hello World"
# 访问字符串中的第一个字符
print(s[0]) # H
# 访问字符串中的最后一个字符
print(s[-1]) # d
# 访问字符串中的一部分（切片）
print(s[0:5]) # Hello
print(s[6:]) # World


'''
小括号()用于创建元组、改变运算优先级以及表示函数调用。
'''
print('-' * 18, "小括号()用于创建元组或改变运算优先级或表示函数调用", '-' * 18)
# 元组是一种有序的，不可变的序列类型，用于存储任意类型的元素。
# 创建一个元组（注意：单个元素的元组需要加逗号）
t = (1, 2, 3)
u = (4,) 
v = () # 空元组
# 访问元组中的元素（和索引类似）
print(t[0])
print(t[-1])
print(t[0:2])

# 改变运算优先级是指使用小括号来改变表达式中运算符的执行顺序。
# 不使用小括号，按照默认优先级计算（乘除高于加减）
x = 2 + 3 * 4
print(x)
# 使用小括号，改变优先级，先计算加法再计算乘
x = (2 + 3) * 4
print(x)
# 用于函数的定义和调用。
def add(x,y):
  return x+y
print(add(1,2))
```
