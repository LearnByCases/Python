Python中的列表是一种有序，可变，可重复的数据结构，可以用来存储多个数据。列表使用方括号创建，列表中的元素可以是任意类型。

列表的常见操作：
* list.count(x)：返回列表中值为x的元素的个数。
* list.index(x[, start[, end]])：返回列表中第一个值为x的元素的索引，可以指定起始和结束位置来限制搜索范围。
* list.append(x)：在列表末尾添加一个元素x。
* list.pop([i])：从列表中移除并返回第i个位置的元素，如果没有指定i，则移除并返回最后一个元素。
* list.sort(*, key=None, reverse=False)：对列表中的元素进行排序，可以指定排序的关键字和顺序。
* list.reverse()：反转列表中的元素顺序。
* list.extend(iterable)：在列表末尾添加一个可迭代对象的所有元素。
* list.insert(i, x)：在列表的第i个位置插入一个元素x。
* list.remove(x)：从列表中移除第一个值为x的元素。
* list.copy()：返回列表的一个浅拷贝。
* list.clear()：清空列表中的所有元素。

```python
# 创建一个列表
fruits = ['orange', 'apple', 'pear', 'banana', 'apple', 'banana']
fruitsx = ['grape', 'melon']

# 统计'apple'在列表中出现的次数
print('-' * 18, "list.count(x)", '-' * 18)
print(r'fruits.count("apple")')
print(f'{fruits}中apple出现了{fruits.count("apple")}次')
print(f'{fruits}中applex出现了{fruits.count("applex")}次')

# 查找'banana'在列表中第一次出现的位置
print('-' * 18, "list.index(x[, start[, end]])", '-' * 18)
print(r'fruits.index("banana") + 1')
print(f'{fruits}中banana首次出现在第{fruits.index("banana") + 1}个元素的位置')
print(r'fruits.index("banana", 4) + 1')
# 查找'banana'在列表中从第4个位置开始第一次出现的位置
print(f'{fruits}中banana在从第四个位置开始，首次出现在第{fruits.index("banana", 4) + 1}个元素的位置')

# 在列表末尾添加一个元素'grape'
print('-' * 18, "list.append(x)", '-' * 18)
print(r'fruits.append("grape")')
fruits.append("grape")
print(f'追加grape后的fruits：\n{fruits}')

# 从列表中移除并返回最后一个元素
print('-' * 18, "list.pop([i])", '-' * 18)
print(r'fruits.pop()')
print(f'从列表中移除{fruits.pop()}后的fruits：\n{fruits}')

# 对列表中的元素进行排序
print('-' * 18, "list.sort(*, key=None, reverse=False)", '-' * 18)
print(r'fruits.sort()')
fruits.sort()
print(f'排序后的fruits：\n{fruits}')

# 反转列表中的元素顺序
print('-' * 18, "list.reverse()", '-' * 18)
print(r'fruits.reverse()')
fruits.reverse()
print(f'反转后的fruits：\n{fruits}')

# 在列表末尾添加另一个列表的所有元素
print('-' * 18, "list.extend(iterable)", '-' * 18)
print(r'fruits.extend(fruitsx)')
print(f'在{fruits}后追加{fruitsx}后得到：')
fruits.extend(fruitsx)
print(fruits)

# 在列表的第2个位置插入一个元素'cherry'
print('-' * 18, "list.insert(i, x)", '-' * 18)
print(r'fruits.insert(1, "cherry")')
print(f'在{fruits}的第二个元素位置插入cherry后得到：')
fruits.insert(1, "cherry")
print(fruits)

# 从列表中移除第一个值为'banana'的元素
print('-' * 18, "list.remove(x)", '-' * 18)
print(r'fruits.remove("banana")')
print(f'从{fruits}中移除第一个值为banana的元素后得到：')
fruits.remove("banana")
print(fruits)

# 修改列表中第一个元素
print('-' * 18, "修改列表中的元素", '-' * 18)
print(r'fruits[0] = "PEAR"')
print(f'将{fruits}第一个元素的值由{fruits[0]}改为PEAR后得到：')
fruits[0] = "PEAR" # 使用索引赋值
print(fruits)
# 修改列表中第二个到第三个元素（不包括第四个）
print(r'fruits[1:3] = ["CHERRY", "ORANGE"]')
print(f'将{fruits}的第二、第三个元素的值由{fruits[1]}、{fruits[2]}改为CHERRY、ORANGE后得到：')
fruits[1:3] = ["CHERRY", "ORANGE"] # 使用切片赋值
print(fruits)

# 返回列表的一个浅拷贝
print('-' * 18, "list.copy()", '-' * 18)
print(r'fruitsy = fruits.copy()')
fruitsy = fruits.copy()
print(f'fruits: {fruits}\n浅拷贝后得到\nfruitsy: {fruitsy}')

# 清空列表中的所有元素
print('-' * 18, "list.clear()", '-' * 18)
print(r'fruits.clear()')
print(f'将{fruits}清空所有元素后得到：')
fruits.clear()
print(fruits)
```

Python中列表遍历的常见方法
* 使用for循环：这是最常用的方法，可以直接用for循环遍历列表中的每个元素，或者用range()函数和len()函数配合遍历列表中的每个索引。
* 使用while循环：这是另一种常用的方法，可以用while循环和一个计数变量来遍历列表中的每个元素或索引。
* 使用enumerate()函数：这是一种更高效的方法，可以用enumerate()函数将列表转换为一个索引序列，同时获取元素的索引和值，一般用在for循环当中。
* 使用列表推导式：这是一种简洁的方法，可以用列表推导式来创建一个新的列表，并对原列表中的每个元素进行操作或筛选。
* 使用lambda函数：这是一种灵活的方法，可以用lambda函数来定义一个匿名函数，并对原列表中的每个元素进行操作或筛选。
* 使用numpy模块：这是一种科学计算的方法，可以用numpy模块中的函数或方法来对原列表进行各种数学或逻辑运算。

```python
# 创建一个列表
mylist = [1, 2, 3]

print('-' * 18, "使用for循环遍历列表中的每个元素", '-' * 18)
print(r'for x in mylist:')
for x in mylist:
    print(x)

print('-' * 18, "使用for循环遍历列表中的每个索引", '-' * 18)
print(r'for i in range(len(mylist)):')
for i in range(len(mylist)):
    print(i, mylist[i])

print('-' * 18, "使用while循环遍历列表中的每个元素", '-' * 18)
print(r'while i < len(mylist):')
i = 0
while i < len(mylist):
    print(mylist[i])
    i += 1

print('-' * 18, "使用while循环遍历列表中的每个索引", '-' * 18)
print(r'while i < len(mylist):')
i = 0
while i < len(mylist):
    print(i, mylist[i])
    i += 1

print('-' * 18, "使用enumerate()函数遍历列表中的每个元素和索引", '-' * 18)
print(r'for i, x in enumerate(mylist):')
for i, x in enumerate(mylist):
    print(i, x)

print('-' * 18, "使用列表推导式创建一个新的列表，并对原列表中的每个元素乘以2", '-' * 18)
newlist = [x * 2 for x in mylist]
print(r'[x * 2 for x in mylist]' + f'\n{newlist}')

print('-' * 18, "使用lambda函数创建一个新的列表，并对原列表中的每个元素加上1", '-' * 18)
newlist = list(map(lambda x: x + 1, mylist))
print(r'list(map(lambda x: x + 1, mylist))' + f'\n{newlist}')
```
