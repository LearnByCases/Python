```python
'''
返回两个字符串的最长不重复子串
'''
# 直接比较
def LNRS1(string1,string2):
  # 获取两个字符串的长度
  len1 = len(string1)
  len2 = len(string2)
  # 初始化一个变量，用来记录最长不重复子串的长度
  max_len = 0
  # 初始化一个变量，用来记录最长不重复子串的内容
  max_str = ""
  # 遍历两个字符串，比较每个子串是否不重复
  for i in range(len1):
    for j in range(i,len1):
      # 获取第一个字符串的子串
      sub1 = string1[i:j+1]
      # 判断第一个字符串的子串是否不重复，即没有相同的字符
      if len(sub1) == len(set(sub1)):
        # 遍历第二个字符串，比较每个子串是否不重复
        for k in range(len2):
          for l in range(k,len2):
            # 获取第二个字符串的子串
            sub2 = string2[k:l+1]
            # 判断第二个字符串的子串是否不重复，即没有相同的字符
            if len(sub2) == len(set(sub2)):
              # 判断两个子串是否相等，即是否是最长不重复子串
              if sub1 == sub2:
                # 如果当前子串长度大于之前记录的最大长度，更新最大长度和最长子串内容
                if len(sub1) > max_len:
                  max_len = len(sub1)
                  max_str = sub1
  # 返回最长不重复子串的长度和内容
  return max_len, max_str

# 动态规划
def LNRS2(X, Y):
  # 获取字符串的长度
  m = len(X)
  n = len(Y)
  # 初始化dp数组
  res = [[0] * (n + 1) for _ in range(m + 1)]
  # 记录最大长度和起始位置
  max_len = 0
  end = 0
  # 遍历两个字符串
  for i in range(1, m + 1):
    for j in range(1, n + 1):
      # 如果字符相等，更新res值
      if X[i - 1] == Y[j - 1]:
        res[i][j] = res[i - 1][j - 1] + 1
        # 如果res值大于最大长度且字符串无重复，更新最大长度和结尾位置
        if res[i][j] > max_len:
          max_len = res[i][j]
          end = i

  # 根据最大长度和结尾位置，截取最长公共子串
    lcs = X[end - max_len : end]
    new_lcs = ''
  # 根据最长公共子串，获得最长不重复公共子串
  for i in range(max_len):
    for j in range(i, max_len + 1):
      if len(lcs[i:j]) == len(set(lcs[i:j])) and len(lcs[i:j]) > len(new_lcs):
        new_lcs = lcs[i:j]
  return len(new_lcs), new_lcs

# 哈希算法
def LNRS3(string1,string2):
  # 获取两个字符串的长度
  len1 = len(string1)
  len2 = len(string2)
  # 初始化一个字典，用来存储字符串的hash值和位置
  hash_dict = {}
  # 初始化一个变量，用来记录最长公共子串的长度
  max_len = 0
  # 初始化一个变量，用来记录最长公共子串的内容
  max_str = ""
  # 遍历较短的字符串，计算每个子串的hash值，并存入字典
  for i in range(len2):
    for j in range(i,len2):
      # 计算子串的hash值
      if len(string2[i:j+1]) != len(set(string2[i:j+1])):
        continue
      hash_value = hash("".join(string2[i:j+1]))
      # 将hash值和子串位置存入字典，如果有重复的hash值，只保留第一个出现的位置
      if hash_value not in hash_dict:
        hash_dict[hash_value] = (i,j)
  # 遍历较长的字符串，计算每个子串的hash值，并在字典中查找是否有匹配的子串
  for i in range(len1):
    for j in range(i,len1):
      if len(string1[i:j+1]) != len(set(string1[i:j+1])):
        continue
      # 计算子串的hash值
      hash_value = hash("".join(string1[i:j+1]))
      # 如果在字典中找到了相同的hash值，说明可能有相同的子串
      if hash_value in hash_dict:
        # 获取较短字符串中对应的子串位置
        start,end = hash_dict[hash_value]
        # 比较两个子串是否完全相等，避免hash冲突的情况
        if string1[i:j+1] == string2[start:end+1]:
          # 如果当前子串长度大于之前记录的最大长度，更新最大长度和最长子串内容
          if j-i+1 > max_len:
            max_len = j-i+1
            max_str = string1[i:j+1]
  # 返回最长公共子串的长度和内容
  return max_len, max_str


# 测试一下函数是否正确
string1 = "abbadefgxy234"
string2 = "xyzabbadef567"
print(LNRS1(string1,string2))
print(LNRS2(string1,string2))
print(LNRS3(string1,string2))
```
